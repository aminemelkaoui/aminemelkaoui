 ## automatisation

```python
def woe_calc(df_auto, descrit_var,good_baad):
    df_auto= pd.concat([df_auto[descrit_var],good_baad],axis=1)
    df_auto=pd.concat([df_auto.groupby(df_auto.columns.values[0], as_index=False)[df_auto.columns.values[1]].count(),
          df_auto.groupby(df_auto.columns.values[0], as_index=False)[df_auto.columns.values[1]].mean()], axis=1)
    df_auto=df_auto.iloc[:,[0,1,3]]
    df_auto.columns=[df_auto.columns.values[0],'n_obs','prop_good']
    df_auto['prop_n_obs']=df_auto['n_obs']/df_auto['n_obs'].sum()
    df_auto['n_good']=df_auto['n_obs']*df_auto['prop_good']
    df_auto['n_bad']=df_auto['n_obs']-df_auto['n_good']
    df_auto['prop_n_good']=df_auto['n_good']/df_auto['n_good'].sum()
    df_auto['prop_n_bad']=df_auto['n_bad']/df_auto['n_bad'].sum()
    df_auto['woe']=np.log(df_auto['prop_n_good']/df_auto['prop_n_bad'])
    df_auto=df_auto.sort_values(['woe'])
    df_auto=df_auto.reset_index(drop=True)
    df_auto['diff_group_good']=df_auto['prop_good'].diff().abs()
    df_auto['diff_woe']=df_auto['woe'].diff().abs()
    df_auto['info_value']=(df_auto['prop_n_good']-df_auto['prop_n_bad'])*df_auto['woe']
    df_auto['info_value']=df_auto['info_value'].sum()
    return df_auto
```

```python
df_essy= woe_calc(df_input_p,'grade',df_target_p)
df_essy
```

```python
def vizapp(woe_data, rota_axis_x=0):
    x=np.array(woe_data.iloc[:,0].apply(str))
    y=woe_data['woe']
    plt.figure(figsize=(18,8))
    plt.plot(x, y, marker='o', linestyle='--', color='K')
    plt.xlabel(woe_data.columns[0])
    plt.ylabel('weight of evidence')
    plt.title(str('weight of evidence ')+woe_data.columns[0])
    plt.xticks()
```

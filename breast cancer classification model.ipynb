```python
DF_x=df_cancer.drop(['target'],axis=1)
DF_x
```

```python
DF_y=df_cancer['target']
DF_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(DF_x, DF_y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cancer_svc_model= SVC()
```

```python
cancer_svc_model.fit(X_train, y_train)
```

```python
y_pred1= cancer_svc_model.predict(X_test)
```

```python
y_pred1
```

```python
conf_metr1=confusion_matrix(y_test,y_pred1)
```

```python
sns.heatmap(conf_metr1, annot=True)
```

## nice model now normilzing and improving

```python
print(classification_report(y_test, y_pred1))
```

```python
min_train=X_train.min()
```

```python
rngtrain= (X_train-min_train).max()
```

```python
X_train_norm=(X_train-min_train)/rngtrain
```

```python
X_train_norm
```

```python
min_test=X_test.min()
rngtest= (X_test-min_test).max()
X_test_norm=(X_test-min_test)/rngtest
```

```python
cancer_svc_model.fit(X_train_norm, y_train)
```

```python
y_new_pred= cancer_svc_model.predict(X_test_norm)
```

```python
cm=confusion_matrix(y_test,y_new_pred)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_new_pred))
```
